```python
# Loading the CSV data into a DF
df = pd.read_csv("/home/ainz/Documents/100 School/105 Thesis/Data/Barangay Balibago Facilities Energy Consumption and Appliance Usage Survey (Responses) - Form Responses 1.csv")

# Display the Dataframe
df.head()
```

```python
# Renaming Columns
new_title = ['Timestamp', 'Name', 'Facility', 'Energy Source', 'Daily Elec Use (hr)', 'Monthly Bill', 'Elec Source Rating',\
             'Power Outage', 'OUTAGE: # of DD in a MM', 'OUTAGE: # of HH in a DD',\
             'Alt Power Source',\
             'LIGHT: Type', 'LIGHT: Count', 'LIGHT: # of DD per WK', 'LIGHT: # of HH per DD',\
             'FAN: Type', 'FAN: Count', 'FAN: # of DD per WK', 'FAN: # of HH per DD',\
             'TV: Type', 'TV: Size', 'TV: Count', 'TV: # of DD per WK', 'TV: # of HH per DD',\
             'RADIO: Own', 'RADIO: Type', 'RADIO: Count', 'RADIO: # of DD per WK', 'RADIO: # of HH per DD',\
             'PC: Own', 'PC: Count', 'PC: # of DD per WK', 'PC: # of HH per DD',\
             'LT: Own', 'LT: Size', 'LT: Count', 'LT: # of DD per WK', 'LT: # of HH per DD',\
             'CP: Own', 'CP: Count', 'CP: # of DD per WK',\
             'INTRNT: Own', 'INTRNT: Count', 'INTRNT: # of DD per WK', 'INTRNT: # of HH per DD',\
             'Light', 'Fan', 'TV', 'Radio', 'PC/LT', 'Phone', 'Internet',\
             'Interest on Alt Power Source', 'Not Intersted on Alt Power Source']

try:
    print(successMessage)
    df = df.set_axis(new_title, axis=1)
except KeyError:
    print(errorMessage)

df.head()
```

```python
# Dropping all unecessary columns
columns_to_drop = ['Timestamp', 'Name']

total_rows = df.shape[0]
null_counts = df.isnull().sum()

df.head()


all_null_columns = df.columns[null_counts == total_rows]
columns_to_drop.extend(all_null_columns.tolist())

try:
    print(successMessage)
    df.drop(columns_to_drop, axis=1, inplace=True)
    
except KeyError:
    print(errorMessage)


df.head()
```

```python
# Rankings Dataframe
rankings_df = df[['Light', 'Fan', 'TV', 'Radio', 'PC/LT', 'Phone', 'Internet']]
ranks_map = {'1st': 1, '2nd': 2, '3rd': 3, '4th': 4, '5th': 5, '6th': 6, '7th': 7}

def convert_to_numerical(column):
    return column.replace(ranks_map)

try:
    print(successMessage)
    rankings_df = rankings_df.apply(convert_to_numerical, axis=1)
except KeyError:
    print(errorMessage)

rankings_df.head()
```

```python
ranks = []
for i in range(0, len(rankings_df.columns)):
  ranks.append(i + 1)

percent_position = []
for num in ranks:
    value = 100 * (num - 0.5) / len(ranks)
    percent_position.append(round(value, 2))

print(percent_position)

# Using the Garrett Score Lookup Table
garrett_score_value = {1: 79, 2: 66, 3: 57, 4: 50, 5: 43, 6: 34, 7: 21}

rankings_df = rankings_df.replace(garrett_score_value)

rankings_df.head()
```

```python
# Function for Bar Graph
def showBarGraph(df, title, x_label, y_label, barLabel_decimal):
    df.plot(kind="bar")
    for bar in plt.gca().containers[0]:
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), round(bar.get_height(), barLabel_decimal), ha='center', va='bottom')

    try:
        plt.xlabel(x_label)
        plt.ylabel(y_label)
        plt.title(title)
        plt.show()
    except e:
        return e
        
    return
```

```python
# Ranking of Appliances Importance
ranks_mean = rankings_df.mean()
print(ranks_mean)

ranks_mean_sorted = ranks_mean.sort_values(ascending=False)

showBarGraph(ranks_mean_sorted, "Importance Ranking", "Score", "Appliance", 2)

top_5_appliances = ranks_mean_sorted.nlargest(5)

print("Top 5 Appliances (Highest to Lowest):")
print(top_5_appliances)
```

```python
# Interpreation of Energy Source Satisfaction Likert
energy_likert_df = df['Elec Source Rating']

def interpret_likert(df):
    mean = round(df.mean(), 1)
    if mean >= 1 and mean <= 1.8:
        interpretation = "Very Dissatisfied"
    elif mean >= 1.9 and mean <= 2.6:
        interpretation = "Dissatisfied"
    elif mean >= 2.7 and mean <= 3.4:
        interpretation = "Neither"
    elif mean >= 3.5 and mean <= 4.2:
        interpretation = "Satisfied"
    else:
        interpretation = "Very Satisfied"

    return print(f'Mean: {mean} => {interpretation}')
    
interpret_likert(energy_likert_df)
```
